 ### Q1: Create a list named my_list.

```python
my_list = [45.4, 44.2, 36.8, 35.1, 39.0, 60.0, 47.4, 41.1, 45.8, 35.6]

# a. print the 5th element in the list.
my_list[4] 
```

```python
# b. append 55.2 to my list.
my_list.append(55.2) 
my_list
```

```python
# c. remove the 6th element in the list.
del(my_list[5]) 
my_list   
```

```python
# d. iterate over the list to print data points greater than 45.
for i in my_list:
    if i > 45:
        print(i)
```

### Q2: Introduction to numpy. 

```python
# a. import numpy library
import numpy as np

# b. declare numpy array with the same data points as in my_list.
my_list = np.array(my_list)
my_list
```

```python
#e. compute the mean and std of the array.
a = np.mean(my_list)
print ("Mean is %.2f." %a)
b = np.std(my_list)
print ("Std is %.2f." %b)
```

```python
#d. use logical referencing to get only those values that are less than 45.
my_list2 = my_list[np.where(my_list < 45)]
my_list2
```

```python
#e. compute the max and min of the array.
c = np.max(my_list)
print ("Maximun is %.1f." %c)
d = np.min(my_list)
print ("Minimum is %.1f." %d)
```

### Q3: Introduction to pandas. 

```python
# a. import pandas library.
import pandas as pd

# b. read the IRIS dataset into iris.
iris = pd.read_csv('iris.csv')
```

```python
# c. display the head of the dataset.
iris.head()
```

```python
# d. drop the id column.
iris_new = iris.drop('Id', axis=1)
iris_new.head()
```

```python
# e. subset dataframe to create a new data frame that includes only the measurements for the setosa species.
sub1 = iris.loc[iris['Species'] == 'Iris-setosa']
sub1.head()
```

```python
# f. get the summary statistics.
sub1.describe()
```

```python
#g. create grouped data frames by Species and compute summary statistics.
iris_new2 = iris.groupby("Species")
iris_new2.describe()
```

```python
# h. plot boxplots by Species.
%matplotlib inline
import matplotlib.pyplot as plt

# plot "SepalLengthCm" by Species.
iris.boxplot(column="SepalLengthCm", by="Species")
plt.suptitle("")
plt.show()
```

```python
# plot "SepalWidthCm" by Species.
iris.boxplot(column="SepalWidthCm", by="Species")
plt.suptitle("")
plt.show()
```

```python
# plot "PetalLengthCm" by Species.
iris.boxplot(column="PetalLengthCm", by="Species")
plt.suptitle("")
plt.show()
```

```python
# plot "PetalWidthCm" by Species.
iris.boxplot(column="PetalWidthCm", by="Species")
plt.suptitle("")
plt.show()
```

```python
# i. Plot a scatter matrix plot using the seaborn library.
import seaborn as sns
sns.pairplot(iris,hue='Species')
plt.show()
```
